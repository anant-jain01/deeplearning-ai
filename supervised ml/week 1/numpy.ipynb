```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train=np.array([1,2])
y_train=np.array([300,500])
m=len(x_train)
```

```python
plt.scatter(x_train,y_train,marker='x')
```

```python
w=100
b=100
```

```python
def output(x,w,b):
  m=len(x)
  f_wb=np.zeros(m)
  for i in range(m):
    f_wb[i]=w*x[i]+b
  return f_wb
```

```python
temp=output(x_train,w,b)
plt.scatter(x_train,y_train,c='blue',label='actual')
plt.plot(x_train,temp,c='red',label='prediction')
plt.legend()
```
